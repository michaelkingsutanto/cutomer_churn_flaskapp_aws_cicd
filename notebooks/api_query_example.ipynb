```python
import requests

website = 'http://localhost:5000'
```

# 1 Make prediction via HTTP POST API

```python
url = f'{website}/api/predict_one'
model_inputs = {
    'contract_type': 'Month-to-Month',  
    'num_referrals': 0, 
    'num_dependents': 0, 
    'internet_type' : 'Fiber Optic',
    'age': 20,
    'tenure_months': 12,
    'zip_code': '90001',
    'payment_method': 'Credic Card',
    'total_monthly_fee': 20,
    'married': 'No',
}

response = requests.post(url, json=model_inputs)
print(response)  # 200 means OK, 400 means Bad Request
print(response.text)
```

# 2 Make multiple predictions with one API call

```python
import pandas as pd

dataset = pd.read_csv('../data/test_data_for_API.csv', keep_default_na=False, na_values='')
X = dataset.drop(columns='churn_label')
y = dataset['churn_label']

X
```

```python
records = X.to_dict('records') 
records[:3]
```

```python
url = f'{website}/api/predict'

response = requests.post(url, json = records)
print(response)             # 200 means OK, 400 means Bad Request
print(response.text[:150])  # print first 150 characters of response
```
